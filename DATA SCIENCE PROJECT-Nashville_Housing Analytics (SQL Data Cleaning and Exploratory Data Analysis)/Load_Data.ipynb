```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import pyodbc
```

```python
df = pd.read_excel('Nashville_Housing.xlsx')
df.head()
```

```python
server = 'SQL/ServerName'
database = 'DataBaseName'
user = 'User Name'
password = 'Password'

# Define the connection string for SQL Server
connection_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={user};PWD={password}'
```

```python
connection_str = f'mssql+pyodbc://{user}:{password}@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server'

try:
    engine = create_engine(connection_str)
    
    df.to_sql(name = 'Nashville_Housing', con = engine, if_exists = 'replace', index = False)
    
    print("Data export to SQL Server database is successful")
except Exception as e:
    print("Error connecting to SQL Server database: ", str(e))
```
